```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../data/dengue_2301901.csv')

# Converter data e ordenar
df['data_ini'] = pd.to_datetime(df['data_ini_se'])
df = df.sort_values('data_ini')

# Criar coluna de Taxa de Incidência (exemplo hipotético de população)
populacao = 65000 
df['taxa_incidencia'] = (df['casos'] / populacao) * 100000
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='data_ini', y='casos', color='red')
plt.title('Evolução Semanal de Casos de Dengue')
plt.xlabel('Data')
plt.ylabel('Número de Casos')
plt.grid(True)
plt.show()
```
